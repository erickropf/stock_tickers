```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('./history_60d.csv')
df['date']=pd.to_datetime(df['date'])
df.set_index('date', inplace = True)
```

```python
plt.figure(figsize=(30,10))
stock = input("Enter a ticker symbol:")
stock = stock.upper()
stock_df = df[df['symbol'] == stock ]
plt.plot(stock_df['close'].pct_change().rolling(10).std(),label='Volatility');
plt.plot(stock_df['close'],label='Stock Price');
plt.legend(loc='lower right',fontsize=20);
plt.title('Short Term Price History and Volatility',fontsize=30);
plt.xlabel('Daily Closing Prices',fontsize=20);
plt.ylabel('Price in dollars',fontsize=20);
```
